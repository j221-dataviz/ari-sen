```python
import pandas as pd
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('data/CSVs/Created/tweet_data.csv')
df = df.sample(frac = 1)
```

```python
ari = []

for i in df['content']:
    print(f'Is this tweet threatening? \n {i} \n')
    x = input()
    if x == 'y':
        ari.append(1)
    else:
        ari.append(0)
```

```python
df['ari'] = ari
df.head()
```
